```python


url = "https://atlas.ecdc.europa.eu/public/index.aspx"
path = "chromedriver.exe"
options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
#options.add_argument('--incognito')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
#driver = webdriver.Chrome(path, options=options)
s = Service(path)
driver = webdriver.Chrome(service=s)



try:
    # Abra a página da web
    driver.get(url)

    dropdown_saude = driver.find_element(By.CLASS_NAME, 'ecdc-select-container-healthTopic')
    time.sleep(5)
    # Encontre o item "Influenza" usando JavaScript
    item_influenza = driver.find_element(By.XPATH, "//input[@value='29']")  # O valor '29' corresponde a Influenza

    # Execute um clique usando JavaScript
    driver.execute_script("arguments[0].click();", item_influenza)
    
    #driver.find_element(By.ID, 'ecdc-populations')
    
    # Aguarde o menu suspenso de Subpopulação estar visível
    dropdown_subpopulacao = WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, "//button[@title='Subpopulation']"))
    )

    # Clique no menu suspenso de Subpopulação
    dropdown_subpopulacao.click()

    # Aguarde o item "Influenza activity" estar visível
    item_influenza_activity = WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, "//label[text()='Influenza activity']"))
    )

    # Clique no item "Influenza activity"
    item_influenza_activity.click()
    
    time.sleep(5)
    
    
    # Aguarde o botão do menu suspenso estar clicável
    botao_dropdown = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//button[@title='Indicator']"))
    )

    # Clique no botão do menu suspenso para abrir o menu
    botao_dropdown.click()

    # Aguarde o item "Intensity" estar clicável
    item_intensity = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//label[text()='Intensity']"))
    )

    # Clique no item "Intensity"
    item_intensity.click()
    
    
    # Aguarde a janela de bloqueio desaparecer
    elemento_bloqueio = WebDriverWait(driver, 10).until(
        EC.invisibility_of_element_located((By.CLASS_NAME, "blockUI"))
    )

    # Clique no botão "Load data"
    botao_load_data = WebDriverWait(driver, 30).until(
        EC.element_to_be_clickable((By.ID, "ecdc-btn-refresh"))
    )
    botao_load_data.click()
    
    
    time.sleep(5)
    
    botao_export = WebDriverWait(driver, 50).until(
        EC.element_to_be_clickable((By.XPATH, "//button[@title='Export']"))
    )
    
    botao_export.click()
    
    
    time.sleep(5)
    
    # Aguarde a janela de bloqueio desaparecer
    elemento_bloqueio = WebDriverWait(driver, 10).until(
        EC.invisibility_of_element_located((By.CLASS_NAME, "blockUI"))
    )

    # Aguarde até que a modal esteja invisível
    modal_export = WebDriverWait(driver, 10).until(
        EC.invisibility_of_element_located((By.CLASS_NAME, "modal-content"))
    )

    # Clique no botão "Export"
    botao_e = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.ID, "ecdc-btn-export-csv"))
    )
    
    driver.execute_script("arguments[0].click();", botao_e)
    

    # Opcionalmente, você pode realizar outras ações aqui se necessário.
    #soup = BeautifulSoup(driver.page_source, 'html.parser')

finally:
    time.sleep(5)
    # Feche a janela do navegador
    driver.quit()


```
